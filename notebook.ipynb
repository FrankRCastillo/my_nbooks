```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "GOOGL.csv"
df = pd.read_csv(filename, parse_dates=['Date']).head(50)
```

```python
df.head()
```

```python
# Create subplots
fig, (prc_ax, vol_ax) = plt.subplots( 2, 1, figsize = (12, 8), sharex  = True, gridspec_kw = {'height_ratios': [3, 1]})

# Price
prc_ax.plot(df['High']      , label='High'      , color = 'green' , linestyle='--', linewidth = 1)
prc_ax.plot(df['Low']       , label='Low'       , color = 'red'   , linestyle='-' , linewidth = 1)
prc_ax.plot(df['Open']      , label='Open'      , color = 'green' , linestyle='--')
prc_ax.plot(df['Close']     , label='Close'     , color = 'green' , linestyle='-' )
prc_ax.plot(df['Adj Close'] , label='Adj Close' , color = 'green' , linestyle=':' )

prc_ax.set_ylabel('Price')
prc_ax.set_title('Stock Prices Over Time')
prc_ax.legend()
prc_ax.grid(True)

# Volume
vol_ax.bar(df.index, df['Volume'], color='gray', width=1)
vol_ax.set_ylabel('Volume')
vol_ax.set_xlabel('Date')
vol_ax.grid(True)

plt.tight_layout()
plt.show()
```
